```python
import matplotlib.pyplot as plt
from fbprophet import Prophet #Facebook prophet model for time series prediction
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
df.describe()
```

```python
df=df[["Date","WVHT_m"]] #select date and close price for fitting the model
df.columns=['ds','y'] #rename the selected columns
```

```python
fig = plt.figure(figsize=(9, 7), dpi=150)
plt.plot(df['ds'],df['y'])
plt.xlabel('Date')
plt.ylabel('Wave height')
plt.show()
```

```python
#model = Prophet(seasonality_mode='multiplicative',holidays=promo)
#model.fit(data)
model = Prophet(interval_width=0.95,
                growth = 'linear',seasonality_mode='multiplicative', yearly_seasonality=10, daily_seasonality=False)
#model = Prophet(holidays=promo,yearly_seasonality=10,seasonality_mode='multiplicative')
model.add_seasonality(name='monthly', period=30.5,fourier_order=5 )
model.fit(df)
```

```python
future=model.make_future_dataframe(periods=365*24, freq='h') #prediction for number of days in future
```

```python
prediction=model.predict(future)
model.plot(prediction) #plotting

plt.title("Wave height forecasting")
plt.xlabel("date")
plt.ylabel("Wave Height")
plt.show()
```

```python
from fbprophet.plot import add_changepoints_to_plot
fig=model.plot(prediction)
plt.show()
```

```python
a=add_changepoints_to_plot(fig.gca(), model, prediction)
plt.show()
```
